```python

cd_pivdf = cdescribe(pivoted_df).transpose().reset_index()

print(cd_pivdf.columns.values.tolist())
print(cd_pivdf.shape)
```

```python

pct90symlist = cd_pivdf[cd_pivdf.missing_percent <= 5]['symbol'].tolist()

print(type(pct90symlist))
print(len(pct90symlist))
print(pct90symlist)
```

```python

colsnmiss = [col for col in pivoted_df.columns.tolist() if col in pct90symlist]
print(colsnmiss)

# Calculate correlation matrix
correlation_matrix = pivoted_df[colsnmiss].corr()

# Create a mask to remove the diagonal
# mask = np.triu(np.ones_like(correlation_matrix, dtype=bool), k=1)
# mask = np.tril(np.ones_like(correlation_matrix, dtype=bool), k=1)
mask = np.tril(np.ones_like(correlation_matrix, dtype=bool), k=-1)


correlation_matrix_no_diag = correlation_matrix.where(mask)

# Create heatmap
fig = go.Figure(data=go.Heatmap(
                z=correlation_matrix_no_diag.values,
                x=correlation_matrix.columns,
                y=correlation_matrix.index,
                colorscale='RdBu_r',
                zmin=-1, zmax=1))

fig.update_layout(title='Correlation Heatmap of Symbol Targets (Excluding Diagonal)',
                  width=800, height=800,
                  xaxis_showgrid=False,
                  yaxis_showgrid=False,
                  yaxis_autorange='reversed')

fig.show()
```
